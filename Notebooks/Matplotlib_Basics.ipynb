```python
#多图形的绘制 （创建图窗后 画图命令仅仅在此代码块中有效）
x = [1,2,3,4,5]
y1 = [5,4,3,2,1]
y2 = [0,0,0,0,0]
y3 = [12,4,5,3,1]
fig2 = plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
#绘制子图
fig = plt.figure()
plt.subplot(1,3,1),plt.plot(x,y1)
plt.subplot(1,3,2),plt.plot(x,y2)
plt.subplot(1,3,3),plt.plot(x,y3)
#subplot(行数，列数，第i幅图)
```

```python
#网格图 需要一个矩阵

```

```python
#设置颜色 风格 粗细(0.5-3合适) 标记
fig = plt.figure()
plt.plot(x,y1,color = '#F6675D',linestyle = '-.',linewidth = 2,marker = 's')

#marker: .  o  ^  s  D
```

```python
#网格图
import numpy as np
x = np.linspace(0,10,100)
I = np.sin(x) * np.cos(x).reshape(-1,1)
fig1 = plt.figure()
plt.imshow(I)
```

```python
#统计图：直方图 （与条形图bar不同  直方图hist是为了看清某分布的均值与标准差）
data = np.random.randn(1000)
fig = plt.figure()
plt.subplot(1,2,1),plt.hist(data,bins = 30,alpha = 0.5,edgecolor = "#FFFFFF") #bins : 区间划分数量   alpha : 透明度
plt.subplot(1,2,2),plt.hist(data,histtype = "stepfilled") # histtype : 图表类型 
```

```python
#设置坐标轴
x = [1,2,3,4,5,6]
y = [1,3,0,122,3,4]
fig = plt.figure()
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,120)

# plt.axis([0,10,0,120])
# plt.axis('equal') 设置x,y轴等长
```

```python
#标题与轴名称
plt.title('A')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#展示图例
x = [1,2,3,4,5]
y1 = [2,3,4,5,6]
y2 = [10,9,89,8,3]
fig = plt.figure()
plt.plot(x,y1,label = 'a')
plt.plot(x,y2,label = 'b')
plt.legend(frameon = False) #frameon = false去除图例边框  loc :"upper center lower" "left center right" 图例位置 也可为'best'
# ncol 图例列数 默认为1
```
