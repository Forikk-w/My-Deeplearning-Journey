```python
import numpy as np
```

```python
# numpys数组只容纳一种数据类型 分为整数型和浮点型即可
arr1 = np.array([1,2,3.0]) # 内含浮点数即为浮点型数组
print(arr1)
```

```python
#往浮点型数组内插入整数 整数上升为浮点数
#往整数型数组内插入浮点数 浮点数被截断为整数
arr1 = np.array([1,2,4,3])
arr1[0] = 100.9
print(arr1)
```

```python
arr1 = np.array([1,2,3])
arr2 = arr1.astype(float) #标准转换类型方法
print(arr2)
```

```python
arr1 = np.array([1,2,3])
print(arr1 / 1) #遇到除法会上升为浮点类型
```

```python
#数组的形状参数 [1 2 3] 形状为 3 / (3,)
#            [[1 3]] 形状为  (1,2)
#            [[[1 2 3]]]形状为 (1,1,3)
arr1 = np.ones((1,1,3)) #传入形状参数
print(arr1)

print(arr1.shape)
```

```python
# reshape转换维度 传入-1会自动计算
arr1 = np.arange(10)
arr2 = arr1.reshape(2,-1) #需要能整除
print(arr2)
#降级为一维数组
print(arr2.reshape(-1))
```

```python
#一维 ： 向量    二维 ： 矩阵
import numpy as np
```

```python
#创建递增数组 arange  array_range
arr1 = np.arange(0,10,2)
print(arr1)
print(arr1[[0,1]])
```

```python
#创建同值数组
arr1 = np.zeros((1,2,3))
print(arr1)
```

```python
#创建随机数组 np.random
arr1 = np.random.random(5)  #0-1分布
print(arr1)
arr2 = np.random.randint(10,100,(1,5)) #范围[10,100) 整数型随机数组
print(arr2)
arr3 = np.random.normal(0,1,(2,3)) #服从正态分布 输入（均值，标准差，形状）
print(arr3)
```

```python
#访问数组元素
arr = np.array([[1,2,3] ,[4,5,6]])
print(arr[1,1])
```

```python
#花式索引
arr[[1,1],[1,2]] = 100
print(arr)
```

```python
#矩阵切片
arr2 = np.arange(0,20).reshape(4,5)
arr3 = arr2 [0 : 3,1 : -1]
print(arr3)
```

```python
#numpy数组切片仅是视图 不会创建新变量 
#修改切片 原数组同步改变
```

```python
arr = [1,2,3]
copy_arr = arr.copy()
copy_arr[0] = 100;
print(arr,copy_arr)
```

```python
#numpy数组之间赋值也不会创建新变量 仅是原数组的一个绑定
```

```python
#数组的变形
#数组的转置方法.T 仅对矩阵有效 因此遇到向量先转化为矩阵
arr = np.arange(1,4).reshape(1,-1)
arr1 = arr.T
print(arr,arr1)
```

```python
#翻转 np.flipud() np.fliplr  向量只能上下翻转（实际数学中向量是竖着排）
arr1 = np.arange(10)
arr2 = np.flipud(arr1)
print(arr2)
```

```python
#数组的拼接
arr1 = np.array( [[1,2,3] , [4,5,6]] )
arr2 = np.array( [[101,102,103] , [104,105,106] ])
arr3 = np.concatenate([arr1,arr2] ,axis = 1) #axis默认=0（增加行的方向)
print(arr3)
```

```python
#数组的分裂
arr = np.arange(10)
arr1,arr2,arr3 = np.split(arr,[1,5])
print(arr1,arr2,arr3)
```

```python
arr = np.arange(30).reshape(5,6)
print(arr)
arr1,arr2,arr3,arr4 = np.split(arr,[1,2,4],axis = 1)
print(arr1,arr2,arr3,arr4)
```

```python
#数组的运算
#乘法，除法 对应元素直接乘除
arr1 = np.arange(1,10).reshape(3,3)
arr2 = np.arange(11,20).reshape(3,3)
print(arr1 * arr2)
```

```python
#广播（不同形状数组之间运算的情况）
#向量与矩阵 ： 向量升级为行矩阵
#某矩阵是行矩阵或列矩阵 广播以适配另一个矩阵的形状
```

```python
#数组的函数
arr1 = np.arange(5)
arr2 = np.arange(15).reshape(5,3)
print(arr1,'\n')
print(arr2,'\n')
print(np.dot(arr1,arr2))
#np.abs np.sin np.cos ... 指数函数：np.exp() 对数：np.log() 默认ln
```
